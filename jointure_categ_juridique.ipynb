```python
conduck = duckdb.connect(database=':memory:', read_only=False)
source_duck = conduck.read_parquet('../../parquet_demonstration.parquet')
transco_t = conduck.read_parquet('./categJuridique.parquet')
```

```python
requete_1 = ''' 
    SELECT siren FROM transco_t as t
    JOIN source_duck as s ON t.siren = s.IdColl'''
df_result1 = conduck.execute(requete_1).fetchdf()
#conduck.close()
print(df_result1.shape)
df_result1.head(5)
```

On va pré_découper 

```python
requete_2 = ''' 
    SELECT DISTINCT 
        siren, 
        s.Idcoll
    FROM transco_t as t
    JOIN source_duck as s ON t.siren = LEFT(s.IdColl, 9)'''
df_result2 = conduck.execute(requete_2).fetchdf()
#conduck.close()
print(df_result2.shape)
df_result2.head(5)
```

```python
requete_3 = '''
    SELECT DISTINCT siren
    FROM transco_t '''
df_requete_3 = conduck.execute(requete_3).fetchdf()
print(df_requete_3.shape)
df_requete_3.head(5)
```

```python
requete_4 = ''' 
    SELECT DISTINCT IdColl
    From source_duck'''
df_requete_4 = conduck.execute(requete_4).fetchdf()
print(df_requete_4.shape)
df_requete_4.head(5)
```

```python
requete_5 = ''' 
 SELECT DISTINCT IdColl 
 FROM source_duck as s
 ANTI JOIN transco_t as t ON t.siren = LEFT(s.IdColl, 9)'''
df_requete_5 = conduck.execute(requete_5).fetchdf()
print(df_requete_5.shape)
df_requete_5.head(5)
```
