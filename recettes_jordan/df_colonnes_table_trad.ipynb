```python
conn = psycopg2.connect(
    dbname="db_v2",
    user="verzochia",
    password="verzochia",
    host="localhost",
    port="5432"
)

cur = conn.cursor()
cur.execute("""
        SELECT table_name, column_name
        FROM information_schema.columns
        WHERE table_schema = 'public'
    """)
tables_columns = cur.fetchall()

cur.close()
conn.close()

data = [(table, column) for table, column in tables_columns]

df = pd.DataFrame(data, columns=['Table', 'Colonnes'])
df
```

```python
df['Table'].drop_duplicates()
```

```python
df = df.sort_values('Table')
df = df[(df['Colonnes'] != 'id') & 
    (~df['Table'].isin(['fonction_chapitre', 'fonction_referentielle',
                        'fonction_compte','nature_compte',
                        'nature_chapitre','transcodage','fonction_compte_mixte']))]
df 
```
